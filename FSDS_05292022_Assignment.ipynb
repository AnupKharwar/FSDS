```python
# q6 : Define a function for all the above problem statememnt  .

def patgen(x,y):
    while(x<y):
        z=1
        while(z<=x):
            print('*',end=" ")

            z=z+1
        print("\n")
        x=x+1

patgen(1,9)
```

```python
# q16 : Write a fun which will take input as a dict and give me out as a list of all the values 
# even in case of 2 level nesting it should work .

def fn1(d):
    l=[]
    for i in d:
        if type(d[i])==dict:
            for j in d[i]:
                l.append(d[i][j])
   
    return l

fn1({'k1':"v1",'k2':"v2",'k3':"v3",'k4':{"a1":"b1","a2":"b2","a3":"b3"}})
```
